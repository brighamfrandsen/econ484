### Using a function

```python
# see if you can print your name right-justified 
# by using the following function

# Hint: enter your name between the Quotes.
def rightJustified(text, chars_per_line=50): # What does 
    """
    Parameters
    ----------
    name           : (string) 
    chars_per_line : (int) The number of characters allowed on the line.
    
    Returns
    -------
    right_justified_line : (string)
    """
    whitespace = " "*chars_per_line          # Why am I doing this?
    return whitespace[:-len(text)]+text      # What's happenening here?
    
my_name = "Bob Dylan"  # <-- enter your name here
rightJustified(my_name)

# Take a minute to mess around and see if you can follow what's happend. 

# Can you get the assumed chars per line to change without directly
# changing the "=50"?
# Hint: rightJustified(my_name, assumed_chars_per_line)

# Can you get the function to through an error?
# Hint: it can be broken by: 1) enter in a string longer than "chars_per_line"
#                            2) passing in a number.
```
